## Take the first 10 reviews texts. Perform word tokenization, lemmatization, part-of-speech tagging. Use Spacy. Each line should have review ID (i.e., line number from the file), token (i.e. word), lemma, and POS tag.

```python
import pandas as pd
import spacy
nlp = spacy.load("en_core_web_sm")
pd.set_option('display.max_colwidth', 200)
```

```python
df=pd.read_json("dataset.txt", lines=True)
df1=df[["reviewText"]].head(10).reset_index()
df1["review_id"]=df1.index+1
df2=df1.drop("index", axis=1)
df2
```

```python
pd.set_option('display.max_rows', 400)
df2["tokens"]=df1[["reviewText"]].applymap(lambda x: nlp(x))
df_tokens=df2[["review_id", "tokens"]]
df_tokens=df_tokens.explode("tokens")
df_tokens["lemma"]=df_tokens[["tokens"]].applymap(lambda x: x.lemma_)
df_tokens["pos"]=df_tokens[["tokens"]].applymap(lambda x: x.pos_)
df_tokens
```
