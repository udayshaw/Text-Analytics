## Take the first 10 review texts. Perform sentence detection using Spacy. Each line should have review ID (i.e., line number from the file) and the sentence itself.

```python
import pandas as pd
import numpy as np
import spacy
nlp = spacy.load("en_core_web_sm")
pd.set_option('display.max_colwidth', 200)
```

```python
df=pd.read_json("dataset.txt", lines=True)
df1=df[["reviewText"]].head(10).reset_index()
```

```python
df1["review_id"]=df1.index+1
df2=df1.drop("index", axis=1)
df2
```

```python
df2["sentences"]=df2[["reviewText"]].applymap(lambda x: list(nlp(x).sents))
```

```python
df3=df2.explode("sentences")
```
